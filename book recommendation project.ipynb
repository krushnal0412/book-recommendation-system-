import numpy as np
import pandas as pd

```python
books=pd.read_csv('Books.csv')
users=pd.read_csv('Users.csv')
ratings=pd.read_csv('Ratings.csv')
```

```python
#popularity based recommend system
```

```python
#merging ratings and books data with respect to isbn as it is available in both
```

```python
ratings_with_name=ratings.merge(books,on='ISBN')
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
```

```python
num_rating_df
```

```python
#now taking the avg ratings by using mean operating
```

```python
avg_rating_df=ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
#now merging num_rating_df and avg_rating_df with respect to Book-Title and assigning it with popularity_df
```

```python
popularity_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
```

```python
popularity_df
```

```python
#we will only keep books with num_rating which is number of ratings more than 100
```

```python
popularity_df=popularity_df[popularity_df['num_ratings']>=100].sort_values('avg_ratings',ascending=False).head(50)
```

```python
#popularity_df is the top 50 books on this platform
```

```python
#now getting books author name its image and other details by merging it with books dataset
```

```python
popularity_df=popularity_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','BookAuthor','Image-URL-M','num_ratings','avg_ratings']]
```

```python
#collaborative recommendate system
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 80
regular_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(regular_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=10
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    #getting index or position of book 
    index=np.where(pt.index==book_name)[0][0]
    similar_books=sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    for i in similar_books:
        print(pt.index[i[0]])
```

```python
recommend("311 Pelican Court")
```
